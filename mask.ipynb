```python
from pandas.io.formats.info import DataFrameInfoPrinter
import pandas as pd

# Read data from Excel file
df = pd.read_excel('input data for masking.xlsx')

# Manipulate the data 

# For ID - multiple the real value with '2' we can use any random number
#df['ID'] = df['ID'] * 2

# For name - using * to hide the char after 3 index
#df['Name']=df['Name'].str[:3].str.ljust(len(df['Name']),'*')
from faker import Faker

fake = Faker()

# Create a new column with masked names
df['Name'] = df['Name'].apply(lambda x: fake.name() if pd.notnull(x) else None)

# Drop the original 'Name' column
#df = df.drop('Name', axis=1)

# Rename the new column to 'Name'
#df = df.rename(columns={'Masked Name': 'Name'})

# For User id - converting alps into upper case and hide the num
df['User ID'] = df['User ID'].apply(lambda x: str(x).upper())

df['User ID']=df['User ID'].str[:3].str.ljust(len(df['User ID']),'#')

# For Age - assigned some value for particular range 
def some(Age):
  
  if Age>20 and Age<30:
    return "brba324"
  elif Age>30 and Age<40:
    return "prbr487"
  elif Age>40 and Age<50:
    return "stth610"
  else:
    return "mohe240"

df['Age']=df['Age'].apply(some)

# For Gender - imported faker to use the fake names by importing


fake = Faker()

# Create a mapping of original values to fake names
mapping = {
    'M': fake.first_name_male(),
    'F': fake.last_name_female(),
    'O': 'Unknown'
    
}

# Replace the values in the 'Gender' column with fake names
df['Gender'] = df['Gender'].replace(mapping)


# Write data back to Excel file
writer = pd.ExcelWriter('output_data.xlsx', engine='openpyxl') 
df.to_excel(writer, index=False)
writer.save()
```
